```python
n_pairings = 150
pairing_list = []
for i in range(n_pairings):
    pairing_id = i
    pairing_list.append(Pairing(pairing_id, None, None, None, None))
```

```python
# create 100 crew members
crew_ids = np.arange(100)
crew_list = []

n_crew = 100
for i in range(n_crew):
    crew_id = i
    seniority = i
    utilities = np.random.rand(len(pairing_list))
    preferences = {p.id: u for p, u in zip(pairing_list, utilities)}

    crew_list.append(CrewMember(crew_id, None, seniority, preferences))
```

```python
senior_rule = SeniorRule()
```

```python
sim = BidlineSimulator(crew_list, pairing_list, senior_rule)
```
